```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
os.chdir("/home/ricardo/estudos/svd")
```

```python
img = cv2.imread("/home/ricardo/estudos/svd/data/dog.jpg")
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
U, S, V = np.linalg.svd(gray_image, full_matrices=True)
S = np.diag(S)
```

```python
# Aproximaes de diferentes ranks
for r in (5, 20, 100):
    Xapprox = U[:, :r] @ S[0:r, :r] @ V[:r, :]
    
    plt.imshow(Xapprox, cmap='gray')
    plt.axis('off')
    plt.title(f'r = {r}')
    
    plt.show()
```

```python
#Ir mostrar os valores singulares em ordem decrescente.
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

#Mostra a proporo cumulativa, indicando quo rpido a soma atinge certo valor percentual significativo da soma total
plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```
