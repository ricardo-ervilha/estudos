```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3 #inclinação verdadeira
```

```python
x = np.arange(-2,2,0.25)
x = x.reshape(-1, 1)
```

```python
b = a*x + np.random.randn(*x.shape) #ruído 
```

```python
U, S, V = np.linalg.svd(x, full_matrices=False)
                        
S_inv = np.linalg.inv(np.diag(S))    
U_transpose = U.T                    
```

```python
print(V.shape)
print(S_inv.shape)
print(U_transpose.shape)
print(b.shape)
```

```python
Xtilde = V @ S_inv @ U_transpose @ b
```

```python
Xtilde
```

```python
plt.plot(x, a*x, color='k', linewidth=2, label='Linha Verdadeira') # Verdadeiro y = a*x

plt.plot(x, b, 'x', color='r', linewidth=2, label='Ruído') # Medidas com ruído adicionado

plt.plot(x, Xtilde * x, '--', color='b', linewidth=4, label='Regressão') # Medidas com ruído adicionado

plt.xlabel('a')
plt.ylabel('b')

plt.grid(linestyle='--')
plt.legend()
plt.show()
```
